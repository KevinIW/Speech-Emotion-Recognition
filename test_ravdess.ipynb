```python
import os
import glob
import torch
import numpy as np
import pandas as pd
import librosa
import soundfile as sf
import matplotlib.pyplot as plt
import seaborn as sns
from transformers import (
    AutoModelForAudioClassification,
    Wav2Vec2FeatureExtractor
)
from collections import Counter, defaultdict

# Configure paths
actor_paths = [
    r"C:\Sem 6\Deep Learning\speech-emotion-recognition-dl-genap-2024-2025\Actor_01",#male
    r"C:\Sem 6\Deep Learning\speech-emotion-recognition-dl-genap-2024-2025\Actor_11",#male
    r"C:\Sem 6\Deep Learning\speech-emotion-recognition-dl-genap-2024-2025\Actor_02",#female
    r"C:\Sem 6\Deep Learning\speech-emotion-recognition-dl-genap-2024-2025\Actor_08",#female
    r"C:\Sem 6\Deep Learning\speech-emotion-recognition-dl-genap-2024-2025\Actor_15",#male
    r"C:\Sem 6\Deep Learning\speech-emotion-recognition-dl-genap-2024-2025\Actor_22"#female
]
model_path = r"C:\Sem 6\Deep Learning\speech-emotion-recognition-dl-genap-2024-2025\checkpoints\emotion_model.pth"
checkpoint = "ehcalabres/wav2vec2-lg-xlsr-en-speech-emotion-recognition"

# Emotion mappings
# Model's 6 emotions
e_labels = {"marah": 0, "jijik": 1, "takut": 2, "bahagia": 3, "netral": 4, "sedih": 5}
inv_map = {v: k for k, v in e_labels.items()}

# RAVDESS emotion mapping (emotion is the 3rd number in filename)
ravdess_emotions = {
    "01": "neutral",
    "02": "calm",
    "03": "happy",
    "04": "sad",
    "05": "angry",
    "06": "fearful",
    "07": "disgust",
    "08": "surprised"
}

# Map RAVDESS emotions to our model's emotions
ravdess_to_model = {
    "neutral": "netral",
    "happy": "bahagia",
    "sad": "sedih",
    "angry": "marah",
    "fearful": "takut",
    "disgust": "jijik",
    # calm and surprised don't map directly
}

def load_model(model_path):
    """Load a trained model from a checkpoint."""
    try:
        # First create model instance with the right architecture
        model = AutoModelForAudioClassification.from_pretrained(
            checkpoint,
            num_labels=6
        )
        # Then load state dictionary into the model
        state_dict = torch.load(model_path, map_location=torch.device('cpu'))
        model.load_state_dict(state_dict)
        print(f"Loaded saved model parameters from {model_path}")
    except Exception as e:
        print(f"Error loading model: {e}")
        # If loading fails, use the base model
        model = AutoModelForAudioClassification.from_pretrained(
            checkpoint,
            num_labels=6
        )
        print("Using base model without fine-tuning")
    
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = model.to(device)
    model.eval()
    return model, device

def extract_emotion_from_filename(filename):
    """Extract emotion from RAVDESS filename format."""
    # RAVDESS filename format: XX-XX-XX-XX-XX-XX-XX.wav
    # The emotion is the 3rd segment (index 2 when split by '-')
    parts = os.path.basename(filename).split('-')
    if len(parts) >= 3:
        emotion_code = parts[2]
        return ravdess_emotions.get(emotion_code, "unknown")
    return "unknown"

def extract_mode_from_filename(filename):
    """Extract mode (speech/song) from RAVDESS filename."""
    parts = os.path.basename(filename).split('-')
    if len(parts) >= 2:
        return "speech" if parts[1] == "01" else "song"
    return "unknown"

def extract_intensity_from_filename(filename):
    """Extract emotional intensity from RAVDESS filename."""
    parts = os.path.basename(filename).split('-')
    if len(parts) >= 4:
        return "normal" if parts[3] == "01" else "strong"
    return "unknown"

def predict_emotion(model, feature_extractor, audio_path):
    """Predict emotion for a single audio file."""
    try:
        # Load and process audio
        speech, sr = sf.read(audio_path)
        if sr != 16000:
            speech = librosa.resample(
                speech, 
                orig_sr=sr, 
                target_sr=16000
            )
        
        if len(speech.shape) > 1:
            speech = np.mean(speech, axis=1)
        
        inputs = feature_extractor(
            speech,
            sampling_rate=16000,
            return_tensors="pt",
            padding=True
        )
        
        # Move inputs to the same device as model
        inputs = {k: v.to(model.device) for k, v in inputs.items()}
        
        # Run inference
        with torch.no_grad():
            outputs = model(**inputs)
            
        # Get predictions
        logits = outputs.logits.cpu().numpy()[0]
        probabilities = torch.nn.functional.softmax(torch.tensor(logits), dim=0).numpy()
        pred_idx = np.argmax(probabilities)
        pred_emotion = inv_map[pred_idx]
        
        # Return all probabilities for detailed analysis
        prob_dict = {inv_map[i]: float(probabilities[i]) for i in range(len(probabilities))}
        
        return pred_emotion, prob_dict
    except Exception as e:
        print(f"Error processing {audio_path}: {e}")
        return "error", {}

def analyze_results(results, title):
    """Analyze and visualize prediction results."""
    # Calculate overall accuracy (only for emotions that map directly)
    correct = 0
    total_mapped = 0
    confusion_matrix = defaultdict(lambda: defaultdict(int))
    
    for result in results:
        true_emotion = result["true_emotion"]
        pred_emotion = result["pred_emotion"]
        
        # Record in confusion matrix
        confusion_matrix[true_emotion][pred_emotion] += 1
        
        # Only count accuracy for directly mappable emotions
        if true_emotion in ravdess_to_model and ravdess_to_model[true_emotion] == pred_emotion:
            correct += 1
            result["is_correct"] = True
        else:
            result["is_correct"] = False
            
        if true_emotion in ravdess_to_model:
            total_mapped += 1
    
    accuracy = correct / total_mapped if total_mapped > 0 else 0
    print(f"\n===== {title} =====")
    print(f"Overall accuracy for mappable emotions: {accuracy:.2%} ({correct}/{total_mapped})")
    
    # Special analysis for calm and surprised
    calm_results = [r for r in results if r["true_emotion"] == "calm"]
    surprised_results = [r for r in results if r["true_emotion"] == "surprised"]
    
    if calm_results:
        calm_predictions = Counter([r["pred_emotion"] for r in calm_results])
        print("\nHow 'calm' samples were classified:")
        for emotion, count in calm_predictions.most_common():
            percentage = count / len(calm_results) * 100
            print(f"  {emotion}: {count} samples ({percentage:.1f}%)")
    
    if surprised_results:
        surprised_predictions = Counter([r["pred_emotion"] for r in surprised_results])
        print("\nHow 'surprised' samples were classified:")
        for emotion, count in surprised_predictions.most_common():
            percentage = count / len(surprised_results) * 100
            print(f"  {emotion}: {count} samples ({percentage:.1f}%)")
    
    # Analysis by intensity (normal vs strong)
    normal_results = [r for r in results if r["intensity"] == "normal"]
    strong_results = [r for r in results if r["intensity"] == "strong"]
    
    if normal_results:
        normal_correct = sum(1 for r in normal_results if r.get("is_correct", False))
        normal_mapped = sum(1 for r in normal_results if r["true_emotion"] in ravdess_to_model)
        normal_accuracy = normal_correct / normal_mapped if normal_mapped > 0 else 0
        print(f"Normal intensity accuracy: {normal_accuracy:.2%} ({normal_correct}/{normal_mapped})")
    
    if strong_results:
        strong_correct = sum(1 for r in strong_results if r.get("is_correct", False))
        strong_mapped = sum(1 for r in strong_results if r["true_emotion"] in ravdess_to_model)
        strong_accuracy = strong_correct / strong_mapped if strong_mapped > 0 else 0
        print(f"Strong intensity accuracy: {strong_accuracy:.2%} ({strong_correct}/{strong_mapped})")
    
    # Visualize results
    create_visualizations(results, title)
    
    return accuracy

def create_visualizations(results, title):
    """Create visualizations for the analysis results."""
    # Set up plotting area with increased size
    plt.figure(figsize=(20, 16))
    plt.suptitle(f"{title} - Emotion Recognition Analysis", fontsize=16, y=0.98)
    
    # 1. Overall accuracy pie chart
    plt.subplot(2, 2, 1)
    mapped_results = [r for r in results if r["true_emotion"] in ravdess_to_model]
    correct = sum(1 for r in mapped_results if r.get("is_correct", False))
    incorrect = len(mapped_results) - correct
    plt.pie([correct, incorrect], labels=["Correct", "Incorrect"], 
            autopct='%1.1f%%', startangle=90, colors=['#4CAF50', '#F44336'])
    plt.title(f"Overall Accuracy\n({correct}/{len(mapped_results)})")
    
    # 2. Calm distribution
    plt.subplot(2, 2, 2)
    calm_results = [r for r in results if r["true_emotion"] == "calm"]
    if calm_results:
        calm_predictions = Counter([r["pred_emotion"] for r in calm_results])
        labels = [f"{e}" for e, c in calm_predictions.most_common()]
        sizes = [c for e, c in calm_predictions.most_common()]
        plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
        plt.title(f"'Calm' Classification\n({len(calm_results)} samples)")
    else:
        plt.text(0.5, 0.5, "No 'calm' samples", ha='center', va='center')
        plt.title("'Calm' Classification")
    
    # 3. Surprised distribution
    plt.subplot(2, 2, 3)
    surprised_results = [r for r in results if r["true_emotion"] == "surprised"]
    if surprised_results:
        surprised_predictions = Counter([r["pred_emotion"] for r in surprised_results])
        labels = [f"{e}" for e, c in surprised_predictions.most_common()]
        sizes = [c for e, c in surprised_predictions.most_common()]
        wedges, texts, autotexts = plt.pie(sizes, labels=None, autopct='%1.1f%%', startangle=90)
        
        # Create legend instead of labels directly on pie
        plt.legend(wedges, labels, title="Emotions", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
        plt.title(f"'Surprised' Classification\n({len(surprised_results)} samples)")
    else:
        plt.text(0.5, 0.5, "No 'surprised' samples", ha='center', va='center')
        plt.title("'Surprised' Classification")
    
    # 4. Confusion matrix heatmap (for mappable emotions)
    plt.subplot(2, 2, 4)
    confusion_data = []
    mappable_emotions = list(ravdess_to_model.keys())
    model_emotions = list(inv_map.values())
    
    for true_emotion in mappable_emotions:
        row = []
        true_count = sum(1 for r in results if r["true_emotion"] == true_emotion)
        for pred_emotion in model_emotions:
            count = sum(1 for r in results if r["true_emotion"] == true_emotion and r["pred_emotion"] == pred_emotion)
            percentage = count / true_count * 100 if true_count > 0 else 0
            row.append(percentage)
        confusion_data.append(row)
    
    df_confusion = pd.DataFrame(confusion_data, index=mappable_emotions, columns=model_emotions)
    sns.heatmap(df_confusion, annot=True, fmt='.1f', cmap='Blues', vmin=0, vmax=100, annot_kws={"size": 9})
    plt.title("Confusion Matrix (% by true emotion)")
    plt.ylabel("True Emotion")
    plt.xlabel("Predicted Emotion")
    
    # Adjust layout to prevent overlapping
    plt.tight_layout()
    plt.subplots_adjust(top=0.92, wspace=0.3, hspace=0.4)
    plt.savefig(f"{title.replace(' ', '_')}_analysis.png", bbox_inches='tight')
    plt.show()

def main():
    # Load model and feature extractor
    print("Loading model and feature extractor...")
    model, device = load_model(model_path)
    feature_extractor = Wav2Vec2FeatureExtractor.from_pretrained(checkpoint)
    
    all_results = []
    
    # Process each actor's directory
    for actor_path in actor_paths:
        actor_id = os.path.basename(actor_path)
        print(f"\nProcessing {actor_id}...")
        
        # Get all wav files
        wav_files = glob.glob(os.path.join(actor_path, "*.wav"))
        print(f"Found {len(wav_files)} audio files")
        
        actor_results = []
        
        # Process each file
        for wav_file in wav_files:
            # Extract information from filename
            true_emotion = extract_emotion_from_filename(wav_file)
            mode = extract_mode_from_filename(wav_file)
            intensity = extract_intensity_from_filename(wav_file)
            
            # Skip song files and only process speech files
            if mode != "speech":
                continue
                
            # Predict emotion
            pred_emotion, probabilities = predict_emotion(model, feature_extractor, wav_file)
            
            # Store result
            result = {
                "filename": os.path.basename(wav_file),
                "true_emotion": true_emotion,
                "pred_emotion": pred_emotion,
                "probabilities": probabilities,
                "mode": mode,
                "intensity": intensity,
                "actor": actor_id
            }
            actor_results.append(result)
            
            # Print progress
            expected_model_emotion = ravdess_to_model.get(true_emotion, "N/A")
            correct = "✓" if expected_model_emotion == pred_emotion else "✗"
            if true_emotion not in ravdess_to_model:
                correct = "?"  # Can't determine correctness for calm and surprised
            
            print(f"File: {os.path.basename(wav_file):30} | True: {true_emotion:10} | Pred: {pred_emotion:10} | {correct}")
        
        # Analyze results for this actor
        analyze_results(actor_results, f"Actor {actor_id}")
        all_results.extend(actor_results)
    
    # Analyze combined results
    analyze_results(all_results, "Combined Results")

if __name__ == "__main__":
    main()
```
